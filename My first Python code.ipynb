# sorting the data in Pandas


```python
df.sort_values(by=["Ticket","Fare"],ascending=[True,False]).head(5)
```

# conditional statement

```python
df["miles_class"] = np.where(df["Fare"] > 50,"high class","low_class")
df.head(10)
```

```python
#multiple conditions 
df["miles_class"] = np.where(df["Fare"] > 60,"high class",np.where(df["Fare"] > 40 ,"Medium_class","low_class"))
df.head(100)
```

# grouping in pandas


```python
df_group = df.groupby("miles_class")["Fare"].agg("sum")
#df_group = df.groupby("miles_class")["Fare"].sum()
df_group
```
