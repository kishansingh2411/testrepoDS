```python
import pandas as pd
import numpy as np
```

```python
url ="https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url,header=None)
#df.head(10)
headers = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
df.columns=headers
df.head(10)
           
```

# Dealing with missing value 

```python
df1=df.replace('?',np.NaN)

df1.dropna(subset=["price"],axis=0,inplace=True)
```

```python
df1["price"].unique()
```

```python
#df1=pd.to_numeric(df1["normalized-losses"])
df1["normalized-losses"]=df1["normalized-losses"].astype('float')
df1["price"] = df1["price"].astype('int')
df1.info()
#mean_normalized_losses = df1["normalized-losses"].mean()
```

```python
mean_normalized_losses = df1["normalized-losses"].mean()
mean_normalized_losses
```

```python
df1["normalized-losses"]=df1["normalized-losses"].replace(np.nan,mean_normalized_losses)
```

```python
df1.head(10)
```

```python
#conver city-mpg to city-L/100KM 
df1["city-mpg"]=235/df1["city-mpg"]
df1.rename(columns={"city-mpg":"city-L/100km"},inplace=True)
```

```python
df1.head(10)
```

```python
df1["price"]=df1["price"].astype("int")
df1.dtypes
```

```python
#normalize length 
#df1["length"]=df1["length"]/df1["length"].max()
```

```python
df1["length"].head(10)
```

```python
#implement bins for price columns to distribute data 
bins = np.linspace(min(df1["price"]),max(df1["price"]),4)
bins
group_names=["low","Medium","High"]
df1["price-binned"]=pd.cut(df1["price"],bins,labels=group_names,include_lowest=True)
df1.hist(by="price-binned",column="price")
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df1["price-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("price")
plt.pyplot.ylabel("count")
plt.pyplot.title("price bins")
```

```python
#convert catagorical variable into int type variable to be feed into models lie fuel column 
fuel_type=pd.get_dummies(df1["fuel-type"],drop_first=True)
fuel_type

dummy_variable_1 = pd.get_dummies(df1["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
df1 = pd.concat([df1, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df1.drop("fuel-type", axis = 1, inplace=True)
```

# Analysis of Variance (ANOVA) 

```python
from scipy import stats
df_Anova = df1[["make","price"]]
grouped_anova = df_Anova.groupby(["make"])
anova_result_1 = stats.f_oneway(grouped_anova.get_group("honda")["price"],grouped_anova.get_group("subaru")["price"])

anova_result_1 
```

```python
from scipy import stats
df_Anova = df1[["make","price"]]
grouped_anova = df_Anova.groupby(["make"])
anova_result_2 = stats.f_oneway(grouped_anova.get_group("honda")["price"],grouped_anova.get_group("jaguar")["price"])

anova_result_2 
```
